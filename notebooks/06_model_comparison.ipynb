```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
results = {
    "Model": [
        "Custom CNN",
        "VGG16",
        "ResNet50 Frozen",
        "ResNet50 Fine-Tuned"
    ],
    
    "Accuracy": [
        84.13,
        85.10,
        80.77,
        82.53
    ],
    
    "Loss": [
        0.5443,
        0.4228,
        0.7885,
        1.0488
    ]
}

comparison_df = pd.DataFrame(results)

comparison_df
```

```python
comparison_df = comparison_df.sort_values(
    by="Accuracy",
    ascending=False
)

comparison_df
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    data=comparison_df,
    x="Model",
    y="Accuracy"
)

plt.title("Model Accuracy Comparison")
plt.ylabel("Accuracy (%)")
plt.xlabel("Model")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    data=comparison_df,
    x="Model",
    y="Loss"
)

plt.title("Model Loss Comparison")
plt.ylabel("Loss")
plt.xlabel("Model")

plt.show()
```

```python
best_model = comparison_df.iloc[0]

print("Best Model:")
print(best_model["Model"])

print("\nAccuracy:")
print(best_model["Accuracy"])

print("\nLoss:")
print(best_model["Loss"])
```
